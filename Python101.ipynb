# Pythonでの四捨五入

## round関数は厳密には四捨五入ではない
## 四捨五入は`Decimal#quantize()`を利用する

```python
from decimal import Decimal, ROUND_HALF_UP, ROUND_HALF_EVEN

target = 0.25

print(str(target),' => round =>', round(target, 1))

dec = Decimal(str(target))

print(str(target),' => quantize(round_half_up) =>', dec.quantize(Decimal('0.1'), rounding=ROUND_HALF_UP))
```

グラフ描画

```python
import pandas_datareader as web

stock_code = '6502.JP'

data_source = 'stooq'

start = '2020-01-01'

end = '2022-05-31'

df = web.DataReader(stock_code, data_source=data_source, start=start, end=end)

print(df)

print(df.shape)

print(df.dtypes)

print(df.size)

df.sort_index(inplace=True)

df.iloc[:,:4].plot()
```
